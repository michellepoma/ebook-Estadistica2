<div style="color: #009929; font-weight: bold;"><h1>La hip√≥tesis nula como un supesto para ser rebatido</h1> </div> 
En el an√°lisis de pruebas de hip√≥tesis, se destaca la hip√≥tesis nula es como una idea inicial que tratamos de cuestionar. No siempre estamos investigando algo nuevo, a veces partimos de la creencia de que una afirmaci√≥n sobre un grupo de cosas es verdadera ,luego usamos la prueba de hip√≥tesis para ver si hay evidencia estad√≠stica que indica lo contrario. En otras palabras es como decir: "Aqu√≠ hay algo que creemos, pero queremos ver si los datos nos dicen que estamos equivocados".

La hip√≥tesis alternativa se formula para desafiar este supuesto (`Œº ‚â† 67.6`). La prueba de hip√≥tesis, basada en una muestra de envases, proporciona evidencia estad√≠stica para aceptar o rechazar la hip√≥tesis nula.  

<div align="center">
$H_0: \mu \geq 67.6$
</div>
<div align="center">
$H_a: \mu < 67,6$
</div>


Si se rechaza la hip√≥tesis nula, se infiere que la afirmaci√≥n del fabricante es incorrecta, respaldando as√≠ la hip√≥tesis alternativa. En este caso, la agencia gubernamental puede tomar medidas para garantizar el cumplimiento de los est√°ndares de etiquetado. Sin embargo, si no se puede rechazar la hip√≥tesis nula, no hay evidencia suficiente para cuestionar la afirmaci√≥n del fabricante, y no se toma ninguna acci√≥n basada en la prueba de hip√≥tesis. Este proceso destaca la importancia de la hip√≥tesis nula como punto de partida para la evaluaci√≥n de afirmaciones y la toma de decisiones basada en evidencia estad√≠stica.

<div style="color: #009929; font-weight: bold;"><h1>Formas para la hipotesis nula y alternativa</h1> </div> 
En las pruebas de hip√≥tesis para la media poblacional, hay tres formas de plantear las hip√≥tesis nula $H_0$ y alternativa $H_a$. Estas se dividen en pruebas de una cola y pruebas de dos colas. Aqu√≠ est√°n las tres formas:

<div align="center">
        <table class= "customTable"> 
             <thead>
            <tr>
                <th>$H_0: \mu \leq \mu_0$</th>
                <th>$H_0: \mu \geq \mu_0$</th>
               <th>$H_0: \mu = \mu_0$</th>
            </tr>
              </thead>
            <tr>
                <td>$H_a: \mu > \mu_0$</td>
                <td>$H_a: \mu < \mu_0$</td>
                <td>$H_a: \mu \neq \mu_0$</td>
            </tr>
        </table>
    </div>
    
Estas formas reflejan la direcci√≥n en la que se est√° interesado en encontrar evidencia significativa. Si se espera que la media poblacional sea mayor que $\mu_0$, se elige la primera forma; si se espera que sea menor, se elige la segunda forma. La tercera forma se selecciona cuando el inter√©s es determinar si hay alguna diferencia significativa, ya sea mayor o menor.

Es importante destacar que la igualdad siempre aparece en la hip√≥tesis nula $H_0$, ya sea con el s√≠mbolo de igualdad $=$.

La elecci√≥n entre estas formas depende de la pregunta que se busca responder y de la direcci√≥n de la evidencia que se espera encontrar. Si se busca respaldar la idea de que la media es mayor, menor o simplemente diferente de $\mu_0$, entonces se elige la forma correspondiente para $H_a$.

Este enfoque permite estructurar adecuadamente las pruebas de hip√≥tesis para la media poblacional y proporciona un marco claro para interpretar los resultados. La comprensi√≥n de estas formas es esencial para evitar errores tipo I y tipo II y garantizar que las pruebas sean efectivas en la evaluaci√≥n de las afirmaciones sobre la media poblacional.

## <div style="display: inline-block; background-color: #009929; padding: 8px; border-radius: 50%; color: white; font-weight: bold;">9.3</div><span style="color:#009929"> Medida Poblacional: &sigma; conocida</span>

En esta secci√≥n se muestra c√≥mo realizar una prueba de hip√≥tesis para la media poblacional en el caso en que œÉ es conocida.

Los m√©todos que se presentan dan resultados exactos si la poblaci√≥n de la que se selecciona la muestra tiene distribuci√≥n normal. En los casos en los que no sea razonable suponer que la poblaci√≥n tiene esta distribuci√≥n, se pueden aplicar estos m√©todos siempre y cuando el tama√±o de la muestra sea suficientemente grande.

A continuaci√≥n se presenta un ejemplo de una prueba para la cola inferior:
<br></br>
El Instituto Boliviano de Regulaci√≥n Comercial (IBRC) realiza estudios para verificar afirmaciones de fabricantes sobre productos, como el peso de caf√© en latas de "Caf√© Sierra".

* La hip√≥tesis nula ($H_0$) establece que la media poblacional del peso de
llenado es al menos 3 libras por lata: $H_0: Œº ‚â• 3$

* La hip√≥tesis alternativa ($H_a$) sugiere que la media es menor de 3 libras: $H_0: Œº < 3$

Se utiliza un nivel de significancia ($Œ± = 0.01$). En el caso de "Caf√© Sierra", si los datos no rechazan $H_0$, no se toman medidas; de lo contrario, se concluye que $H_a$ es verdadera, lo que justificar√≠a un cargo por violaci√≥n en la etiqueta. El director de pruebas de la IBRC est√° dispuesto a asumir un riesgo del 1% de cometer un error tipo I.

Se selecciona una muestra de 36 latas y se calcula la media muestral $x$ como una estimaci√≥n de la media poblacional $Œº$. Si $x$ es menor de 3 libras, se cuestiona la hip√≥tesis nula. El factor clave es el valor de $Œ±$, la probabilidad de cometer un error tipo I. Con los pasos iniciales completados, el siguiente paso es recopilar datos muestrales y calcular el estad√≠stico de prueba para tomar una decisi√≥n basada en la evidencia estad√≠stica.

<div style="color: #009929; font-weight: bold;"><h3>Estad√≠stico de prueba</h3> </div> 
<div style="padding: 10px; background-color: #D4F8B7; color: black; font-size: 1.1em;">
<center>ESTAD√çSTICO DE PRUEBA EN LAS PRUEBAS DE HIP√ìTESIS PARA LA MEDIA POBLACIONAL: œÉ CONOCIDA</center>
$$z = \frac{{\bar{x} - \mu_0}}{{\frac{\sigma}{\sqrt{n}}}}$$
</div>
<br>
</br>
En el estudio de "Caf√© Sierra", se realizan pruebas de hip√≥tesis para la media poblacional ($Œº$) en el caso de que la desviaci√≥n est√°ndar poblacional ($œÉ$) sea conocida. Los resultados previos de la IBRC indican que $œÉ$ tiene un valor de 0.18 y que la distribuci√≥n de los pesos de llenado en la poblaci√≥n sigue una distribuci√≥n normal.

Con un tama√±o de muestra $n = 36$ y asumiendo la hip√≥tesis nula $H_0$ como verdadera $Œº = Œº_0 = 3$, se puede ilustrar la distribuci√≥n de muestreo de la media muestral $\bar{x}$ en una distribuci√≥n normal. El error est√°ndar de $\bar{x}(œÉ_\bar{x})$ se calcula como $\frac{\sigma}{\sqrt{n}} = \frac{0.18}{\sqrt{36}} = 0.03 $

La variable aleatoria normal est√°ndar:

<h4><center>$z =\frac{\bar{x} - \mu_0}{\sigma_{\bar{x}}} = \frac{\bar{x} - 3}{0.03}$</center></h4>

La probabilidad asociada a un valor espec√≠fico de $z$ en la cola inferior se determina utilizando la tabla de probabilidad normal est√°ndar.

Por ejemplo, si $z=-3.00$, el √°rea en la cola inferior es $0.0013$, lo que significa que la probabilidad de obtener un valor de $\bar{x}$ que sea tres o m√°s errores est√°ndar menor que la media hipot√©tica ($\mu_0=3$) es $0.0013$. En una prueba de hip√≥tesis, un resultado tan extremo sugiere que la hip√≥tesis nula es poco probable, lo que podr√≠a llevar al rechazo de la hip√≥tesis nula en favor de la hip√≥tesis alternativa. El estad√≠stico de prueba utilizado es $z$, y se compara con un umbral cr√≠tico basado en el nivel de significancia predefinido ($Œ±$).

<div style="color: #009929; font-weight: bold;">FIGURA 9.1 <div style="color: #5CCB5F; font-weight: bold;">Distribuci√≥n de muestreo de x en el estudio de "Caf√© Sierra" cuando la hip√≥tesis nula es verdadera como igualdad (Œº = 3) </div></div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√°metros de la distribuci√≥n muestral
media_poblacional = 3
error_estandar = 0.03
# Generar datos de la distribuci√≥n muestral
x = np.linspace(media_poblacional - 4 * error_estandar, media_poblacional + 4 * error_estandar, 1000)
y = norm.pdf(x, media_poblacional, error_estandar)

# Crear la gr√°fica
fig, ax = plt.subplots()
ax.plot(x, y, color='#009929')
# Establecer el color de fondo
ax.set_facecolor('#d4f8b7')  
# Pintar el fondo externo del gr√°fico
fig.patch.set_facecolor('#D4F8B7')
ax.set_xlabel(f'Œº = {media_poblacional}', fontsize=12)
ax.set_ylabel(r'Distribuci√≥n de muestreo de $\bar{x}$', fontsize=12)

# Agregar informaci√≥n sobre el error est√°ndar
info_error_estandar = r'$œÉ_\bar{x}$ = $\frac{\sigma}{\sqrt{n}} = \frac{0.18}{\sqrt{36}} = 0.03 $'
ax.text(0.6, 0.85, info_error_estandar, transform=ax.transAxes, fontsize=12, color='#000')

ax.grid(False)
plt.show()
```

La cuesti√≥n clave en una prueba de cola inferior es: **¬øqu√© tan peque√±o debe ser el estad√≠stico de prueba z para que se decida rechazar la hip√≥tesis nula?** Para responder esta pregunta se usandos m√©todos: el m√©todo del **valor-p** y el **m√©todo del valor cr√≠tico**.

<div style="color: #009929; font-weight: bold;"><h3>M√©todo del valor-p</h3> </div>

En este enfoque se usa el valor del estad√≠stico de prueba $z$ paracalcular una probabilidad llamada <strong>valor-p</strong>.

> <div style="padding: 10px; background-color: #D4F8B7; color: black; font-size: 1.1em;"><strong>VALOR-p:</strong> Es una probabilidad que aporta una medida de la evidenciasuministrada por la muestra contra la hip√≥tesis nula. Valores-p peque√±os indican una evidencia mayor contra $H_0$.</div>

<center>El valor - p se utiliza para determinar si la hip√≥tesis nula debe ser rechazada.</center>
<br></br>
El c√°lculo del valor-p depende del tipo de prueba (cola inferior, cola superior o dos colas). Para una prueba de cola inferior con $œÉ$ conocida, se determina el √°rea bajo la curva normal est√°ndar para valores de $z ‚â§$ al valor del estad√≠stico de prueba.

Ahora calculamos el valor-p para la prueba de cola inferior del estudio de <strong>"Caf√© Sierra"</strong>.Suponga que en la muestra de las 36 latas de caf√©, la media muestral obtenida es $\bar{x} = 2.92$ libras. ¬øEs $\bar{x} = 2.92$ lo suficientemente peque√±a para que se rechace $H_0$? Como es una prueba
de cola inferior, el valor-p es el √°rea bajo la curva normal est√°ndar para valores de $z$ que el valor del estad√≠stico de prueba. Al usar $\bar{x} = 2.92$, $œÉ = 0.18$ y $n = 36$, se determina el valor del estad√≠stico de prueba z.

$$z = \frac{{\bar{x} - \mu_0}}{{\frac{\sigma}{\sqrt{n}}}} = \frac{{2.92 - 3}}{{\frac{0.18}{\sqrt{36}}}} = -2.67$$

Por consiguiente, el valor-p es la probabilidad de que el estad√≠stico de prueba $z$ sea menor o igual que $-2.67$(el √°rea bajo la curva normal est√°ndar a la izquierda del estad√≠stico de prueba).

Como se indic√≥ antes, el director del programa de pruebas de la IBRC eligi√≥ como nivel de significancia un valor de 0.01. Seleccionar $Œ± = 0.01$ significa que √©l est√° dispuesto a tolerar una probabilidad de 0.01 para rechazar la hip√≥tesis nula cuando sea verdadera como igualdad ($Œº_0 = 3$). La muestra de 36 latas de "Caf√© Sierra" dio como resultado un $valor-p = 0.0038$, lo cual significa que la probabilidad de obtener $\bar{x} = 2.92$ o menor, si la hip√≥tesis nula considerada como igualdad es verdadera, es 0.0038. Como 0.0038 es menor o igual que $Œ± = 0.01$, $H_0$ es rechazada. De manera que para el nivel de significancia 0.01 se encontr√≥ evidencia estad√≠stica suficiente para rechazar la hip√≥tesis nula.

<div style="color: #009929; font-weight: bold;">FIGURA 9.2 <div style="color: #5CCB5F; font-weight: bold;">Valor-p en el estudio de Hilltop Coffee, en el que $\bar{x} = 2.92$ y $z = -2.67$ </div></div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import t

#GRAFICO1
# Generar datos de la distribuci√≥n muestral
x = np.linspace(0 - 4 * 1, 0 + 4 * 1, 1000)
y = norm.pdf(x*3, 0, 4)

# Crear la gr√°fica
fig, ax = plt.subplots()
ax.plot(x, y, color='#009929')
# Establecer el color de fondo
ax.set_facecolor('#d4f8b7')
# Pintar el fondo externo del gr√°fico
fig.patch.set_facecolor('#D4F8B7')
ax.set_xlabel(f'$Œº_0$ = 3', fontsize=12)

# Agregar informaci√≥n sobre el error est√°ndar
info_error_estandar = r'$œÉ_\bar{x}$ = $\frac{\sigma}{\sqrt{n}} = 0.03 $'
ax.text(0.6, 0.88, info_error_estandar, transform=ax.transAxes, fontsize=13, color='#000')
ax.text(-0.03, -0.15, r'$\bar{x} = 2.92$',transform=ax.transAxes, fontsize=12, color='#000')
ax.text(0.02, 0.85, r'Distribuci√≥n de muestreo',transform=ax.transAxes, fontsize=10, color='#000')
ax.text(0.17, 0.77, r'de $\bar{x}$',transform=ax.transAxes, fontsize=12, color='#000')

# L√≠nea vertical
ax.axvline(t.ppf(0.005, 47), color='#009929', linestyle='--', label=r'$t_{\alpha/2}$')

# Sombrear la parte de -2.67 hacia la izquierda
ax.fill_between(x, y, where=(x <= -2.67), color='#009929', alpha=0.5)


ax.grid(False)
plt.show()

#GRAFICO2
# Generar datos de la distribuci√≥n muestral
x = np.linspace(0 - 4 * 1, 0 + 4 * 1, 1000)
y = norm.pdf(x*3, 0, 4)

# Crear la gr√°fica
fig, ax = plt.subplots()
ax.plot(x, y, color='#009929')
# Establecer el color de fondo
ax.set_facecolor('#d4f8b7')
# Pintar el fondo externo del gr√°fico
fig.patch.set_facecolor('#D4F8B7')

# Agregar informaci√≥n sobre el error est√°ndar
ax.text(0.58, 0.85, r'Distribuci√≥n de muestreo de',transform=ax.transAxes, fontsize=10, color='#000')
ax.text(0.69, 0.72, r'$z = \frac{{\bar{x} - \mu_0}}{{\frac{\sigma}{\sqrt{n}}}}$',transform=ax.transAxes, fontsize=14, color='#000')

ax.text(0.23, 0.07, r'valor-p = 0.0038',transform=ax.transAxes, fontsize=10, color='#000')
ax.text(0.08, -0.15, r'z = -2.67',transform=ax.transAxes, fontsize=12, color='#000')

# L√≠nea vertical
ax.axvline(t.ppf(0.005, 47), color='#009929', linestyle='--', label=r'$t_{\alpha/2}$')

# Sombrear la parte de -2.67 hacia la izquierda
ax.fill_between(x, y, where=(x <= -2.67), color='#009929', alpha=0.5)

ax.grid(False)
plt.show()
```

En la prueba para "Caf√© Sierra", el <strong>valor-p</strong> de 0.0038 condujo al rechazo de la hip√≥tesis nula. Por esta raz√≥n, el valor-p tambi√©n se conoce como nivel de significancia observado.

> <div style="padding: 10px; background-color: #D4F8B7; color: black; font-size: 1.1em;"><strong>El valor cr√≠tico</strong> es el valor del estad√≠stico de prueba que corresponde a un √°rea de Œ± (nivel de significancia) en la cola inferior de la distribuci√≥n de muestreo del estad√≠stico.</div>

En el ejemplo de "Caf√© Sierra", se ilustra el funcionamiento del m√©todo del valor cr√≠tico. En situaciones donde se conoce la desviaci√≥n est√°ndar $œÉ$, la distribuci√≥n de muestreo del estad√≠stico de prueba $z$ es la distribuci√≥n normal est√°ndar. El valor cr√≠tico es aquel que corresponde a un √°rea de $Œ± = 0.01$ en la cola inferior de esta distribuci√≥n.

Consultando la tabla de probabilidad normal est√°ndar, se observa que $z ‚â§ ‚àí2.33$ proporciona un √°rea de 0.01 en la cola inferior (ver figura 9.3). Por lo tanto, si el valor del estad√≠stico de prueba obtenido con la muestra es menor o igual a ‚àí2.33, el valor-p correspondiente ser√° menor o igual a 0.01, lo que lleva al rechazo de la hip√≥tesis nula.
Entonces, en el estudio de "Caf√© Sierra" la regla
para el rechazo usando el valor cr√≠tico para un nivel de significancia de 0.01 es:
<center>Rechazar $H_0$ si el $z ‚â§ -2.33$</center>
<br></br>
En nuestro ejemplo, $\bar{x} = 2.92$ y el estad√≠stico de prueba es $z = -2.67$. Como $z = -2.67 < -2.33$, $H_0$ puede ser rechazada y concluir que "Caf√© Sierra" est√° llenando las latas de manera deficiente.

<div style="color: #009929; font-weight: bold;"><h1>Prueba de dos colas</h1> </div>

En las pruebas de hip√≥tesis, la forma general de una <strong>prueba de dos colas</strong> es la siguiente:

<table style="margin: 0 auto;">
        <tr>
            <td style="background-color: #D4F8B7;">$$H_0: Œº = Œº_0$$</td>
            <td style="background-color: #B7EAB7;">$$H_a: Œº \neq Œº_0$$</td>
        </tr>
    </table>
    
Consideremos el caso de una prueba de hip√≥tesis en la empresa "BoliviaGolfCrafters".

La Federaci√≥n Boliviana de Golf (FBG) establece est√°ndares para los fabricantes de equipos de golf, y "BoliviaGolfCrafters", una empresa de fabricaci√≥n de pelotas de golf, utiliza procesos de alta tecnolog√≠a. La distancia media de recorrido objetivo es 295 yardas, pero el proceso puede desajustarse, produciendo pelotas con distancias diferentes. "BoliviaGolfCrafters" implementa un programa de control de calidad que implica pruebas de hip√≥tesis con muestras de 50 pelotas. La hip√≥tesis nula asume que el proceso est√° funcionando correctamente, con una distancia media de 295 yardas, mientras que la hip√≥tesis alternativa sugiere que la distancia media no es igual a 295 yardas. Estas hip√≥tesis se expresan como:

<center>$H_0: Œº = 295$</center>
<center>$H_a: Œº \ne 295$</center>
<br></br>

Este enfoque permite a "BoliviaGolfCrafters" monitorear y ajustar su proceso de fabricaci√≥n para cumplir con los est√°ndares de la FBG y satisfacer las expectativas de los clientes.

Si la media muestral $\bar{x}$ es significativamente menor o significativamente mayor que 295 yardas, $H_0$ ser√° rechazada. En este caso, se tomar√°n medidas para ajustar el proceso de manufactura. Por otro lado, si $x$ no se desv√≠a una cantidad significativa de la media hipot√©tica $Œº_0 = 295$, $H_0$ no ser√° rechazada, y no se tomar√° medida alguna para ajustar el proceso de manufactura.

El equipo de control de calidad elige $Œ± = 0.05$ como nivel de significancia para esta prueba. Datos de pruebas anteriores realizadas sabiendo que el proceso est√° ajustado, indican que se puede suponer que la desviaci√≥n est√°ndar poblacional se conoce y que su valor es $œÉ = 12$. Por ende, con un tama√±o de muestra $n = 50$, el error est√°ndar $\bar{x}$ es:

<center><h3>$œÉ_\bar{x} = \frac{\sigma}{\sqrt{n}} = \frac{12}{\sqrt{50}} = 1.7 $</h3></center>

Como el tama√±o de la muestra es grande, el teorema del l√≠mite central (cap√≠tulo 7) permite concluir que la distribuci√≥n de muestreo de $\bar{x}$ puede aproximarse mediante una distribuci√≥n normal. En la (figura 9.4) se ilustra la distribuci√≥n de muestreo de $\bar{x}$ para la prueba de hip√≥tesis de "BoliviaGolfCrafters" con una media poblacional hipot√©tica de $Œº_0 = 295$.

<div style="color: #009929; font-weight: bold;">FIGURA 9.4 <div style="color: #5CCB5F; font-weight: bold;">Distribuci√≥n de muestreo de $\bar{x}$ en la prueba de hip√≥tesis de "BoliviaGolfCrafters"</div></div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√°metros de la distribuci√≥n muestral
media_poblacional = 295
error_estandar = 1.7
# Generar datos de la distribuci√≥n muestral
x = np.linspace(media_poblacional - 4 * error_estandar, media_poblacional + 4 * error_estandar, 1000)
y = norm.pdf(x, media_poblacional, error_estandar)

# Crear la gr√°fica
fig, ax = plt.subplots()
ax.plot(x, y, color='#009929')
# Establecer el color de fondo
ax.set_facecolor('#d4f8b7')
# Pintar el fondo externo del gr√°fico
fig.patch.set_facecolor('#D4F8B7')
ax.set_xlabel(f'$Œº_0$ = {media_poblacional}', fontsize=12)

# Agregar informaci√≥n sobre el error est√°ndar
info_error_estandar = r'$œÉ_\bar{x}$ = $\frac{\sigma}{\sqrt{n}} = \frac{12}{\sqrt{50}} = 1.7$'
ax.text(0.63, 0.85, info_error_estandar, transform=ax.transAxes, fontsize=12, color='#000')
ax.text(0.02, 0.85, r'Distribuci√≥n de muestreo',transform=ax.transAxes, fontsize=11, color='#000')
ax.text(0.17, 0.77, r'de $\bar{x}$',transform=ax.transAxes, fontsize=12, color='#000')
ax.text(0.50, 0.01, r'|',transform=ax.transAxes, fontsize=12, color='#000')

ax.grid(False)
plt.show()
```

<div style="color: #009929; font-weight: bold;"><h3>M√©todo del valor-p</h3> </div>

A continuaci√≥n veremos c√≥mo se calcula el <strong>valor-p</strong> en la prueba de hip√≥tesis de "BoliviaGolfCrafters". Primero calculamos el valor del estad√≠stico de prueba. En el caso en que se conoce $œÉ$, el
estad√≠stico de prueba $z$ es la variable aleatoria normal est√°ndar. Empleando la ecuaci√≥n ya conocida con
$\bar{x} = 297.6$, el valor del estad√≠stico de prueba es:

$$z = \frac{{\bar{x} - \mu_0}}{{\frac{\sigma}{\sqrt{n}}}} = \frac{{297.6 - 295}}{{\frac{12}{\sqrt{50}}}} = 1.53$$

En el c√°lculo del valor-p para la prueba de hip√≥tesis de dos colas en "BoliviaGolfCrafters", se busca la probabilidad de obtener un valor tan improbable como $z = 1.53$. Dado que es una prueba de dos colas, se consideran tanto los valores $z ‚â• 1.53$ como $z ‚â§ ‚àí1.53$. La probabilidad se calcula duplicando el √°rea bajo la curva normal est√°ndar a la derecha de $z = 1.53$. Para $z = 1.53$, el √°rea a la izquierda es $0.9370$, y el √°rea a la derecha es $1.0000 - 0.9370 = 0.0630$. Al duplicar este valor, se obtiene $2(0.0630) = 0.1260$ como el <strong>valor-p</strong>.

<div style="color: #009929; font-weight: bold;">FIGURA 9.5 <div style="color: #5CCB5F; font-weight: bold;">Valor-p en la prueba de hip√≥tesis de "BoliviaGolfCrafters"</div></div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar datos de la distribuci√≥n muestral
x = np.linspace(0 - 1 * 2, 0 + 1 * 2, 1000)
y = norm.pdf(x*3, 0, 2)


# Crear la gr√°fica
fig, ax = plt.subplots()
ax.plot(x, y, color='#009929')
# Establecer el color de fondo
ax.set_facecolor('#d4f8b7')
# Pintar el fondo externo del gr√°fico
fig.patch.set_facecolor('#D4F8B7')
ax.set_xlabel(f'$valor-p = 2(0.0630) = 0.1260$', fontsize=10)

# Agregar informaci√≥n sobre el error est√°ndar
ax.text(0.02, 0.15, r'P(z ‚â§ -1.53) = 0.0630',transform=ax.transAxes, fontsize=9, color='#000')
ax.text(0.76, 0.15, r'P(z ‚â• 1.53) = 0.0630',transform=ax.transAxes, fontsize=9, color='#000')

ax.text(0.06, -0.15, r'-1.53',transform=ax.transAxes, fontsize=12, color='#000')
ax.text(0.86, -0.15, r'1.53',transform=ax.transAxes, fontsize=12, color='#000')

# Sombrear la parte de -1.53 hacia la izquierda
ax.fill_between(x, y, where=(x <= -1.53), color='#009929', alpha=0.5)
# Sombrear la parte de 1.53 hacia la derecha
ax.fill_between(x, y, where=(x >= 1.53), color='#009929', alpha=0.5)

ax.grid(False)
plt.show()
```

Al comparar el valor-p con el nivel de significancia $Œ± = 0.05$, se encuentra que $0.1260 > 0.05$. Por lo tanto, la hip√≥tesis nula no es rechazada, indicando que no se necesitan medidas para ajustar el proceso de manufactura de "BoliviaGolfCrafters", ya que la distancia media de recorrido no difiere significativamente de $295$ yardas seg√∫n la muestra analizada.

<div style="padding: 10px; background-color: #D4F8B7; color: black; font-size: 1.1em;">
    <strong>C√ÅLCULO DEL VALOR-p EN UNA PRUEBA DE DOS COLAS</strong>

1. Determine el valor del estad√≠stico de prueba $z$.
2. Si el valor del estad√≠stico de prueba est√° en la cola superior ($z > 0$), encuentre el √°rea bajo la curva normal est√°ndar a la derecha de $z$; si est√° en la cola inferior ($z < 0$), localice el √°rea bajo la curva normal est√°ndar a la izquierda de $z$.
3. Duplique el √°rea, o probabilidad, en la cola, obtenida en el paso 2 y determine el valor-p.
</div>

<div style="color: #009929; font-weight: bold;"><h3>M√©todo del valor cr√≠tico</h3> </div>

En la figura 9.6 se aprecia que los valores cr√≠ticos en esta prueba se encuentran tanto en la cola superior como en la cola inferior de la distribuci√≥n normal est√°ndar. Entonces, al utilizar el m√©todo del valor cr√≠tico, la regla de rechazo para dos colas es:
    
<center>Rechazar $H_0$ si el $z ‚â§ -1.96$ o $z ‚â• 1.96$</center>
<br></br>
Como en el estudio de MaxFlight el valor del estad√≠stico de prueba es $z = 1.53$, la evidenciaestad√≠stica no permitir√° rechazar la hip√≥tesis nula a un nivel de signifi cancia de $0.05$.

<div style="color: #009929; font-weight: bold;"><h1>Ejercicio 1</h1> </div> Considere la prueba de hip√≥tesis siguiente:
<center>$H_0: Œº ‚â§ 25$</center>
<center>$H_a: Œº > 25$</center>
En una muestra de 40, la media muestral es 26.4 y la desviaci√≥n est√°ndar poblacional es 6.

a) Calcule el valor del estad√≠stico de prueba.

b) ¬øCu√°l es el valor-p?

c) Use $Œ± = 0.01$, ¬øcu√°l es su conclusi√≥n?

d) ¬øCu√°l es la regla de rechazo si se usa el m√©todo del valor cr√≠tico? ¬øQu√© concluye?

$Datos:$

$n =40$

$\bar{x} = 26.4$

$œÉ = 6$

$œÉ_\bar{x} = \frac{œÉ}{\sqrt{n}}$



<span style="color: green;"><strong>Inciso a)</strong></span>

$$z = \frac{{\bar{x} - \mu_0}}{{\frac{\sigma}{\sqrt{n}}}} = \frac{{{26.4} - 25}}{{\frac{6}{\sqrt{40}}}} = 1.48$$

<span style="color: green;"><strong>Inciso b)</strong></span>

Se busca el valor de $z = 1.48$ en la tabla normal, entonces:

<center>$valor-p = (1 - 0.9306) = 0.0694$</center>

<span style="color: green;"><strong>Inciso c)</strong></span>

La regla me dice que se debe rechazar $H_0$ si el $valor-p ‚â§ Œ±$, por lo cual tenemos que como el valor-p es mayor que $Œ±$ (0.01) se acepta la hip√≥tesis nula. 

<span style="color: green;"><strong>Inciso d)</strong></span>

El m√©todo del valor critico me dice que debo rechazar $H_0$ si $z ‚â• 2.33$, por lo que tenemos que utilizando el valor de $z$ se acepta la hip√≥tesis nula, ya que $1.48 < 2.33$

<div style="color: #009929; font-weight: bold;"><h1>Ejercicio en c√≥digo</h1> </div>

```python
import math
# Datos
n = 40  # Tama√±o de la muestra
x_bar = 26.4  # Media muestral
sigma = 6  # Desviaci√≥n est√°ndar poblacional

# Estad√≠stico de prueba
mu_0 = 25  # Valor de la hip√≥tesis nula
se = sigma / (n ** 0.5)  # Error est√°ndar de la media
z = (x_bar - mu_0) / se  # C√°lculo del estad√≠stico de prueba (z)
print("Estad√≠stico de prueba (z):", round(z, 2))

# Valor-p
#p_value = 1 - stats.norm.cdf(z)  # C√°lculo del valor-p utilizando la funci√≥n cdf() de la distribuci√≥n normal est√°ndar
p_value = 1 - 0.9306  # C√°lculo del valor-p utilizando la funci√≥n cdf() de la distribuci√≥n normal est√°ndar
print("Valor-p:", round(p_value, 4))

# Conclusi√≥n
alpha = 0.01  # Nivel de significancia
if p_value <= alpha:
    print("Se rechaza la hip√≥tesis nula (H0)")
else:
    print("Se acepta la hip√≥tesis nula (H0)")
    
    
```

$Datos:$

$n = 50$

<div style="color: #009929; font-weight: bold;"><h1>Ejercicio en c√≥digo</h1> </div>

```python
import math
from scipy.stats import norm

# Datos del ejercicio
n = 120
x_barra = 8.5
mu_0 = 8
sigma = 3.2

# C√°lculo del error est√°ndar de la media muestral
sigma_x_barra = sigma / math.sqrt(n)

# C√°lculo del valor del estad√≠stico de prueba (z)
z = (x_barra - mu_0) / sigma_x_barra

# C√°lculo del valor-p (dos colas)
valor_p = (1 - 0.8294)

# Imprimiendo resultados
print(f"Valor del estad√≠stico de prueba (z): {z:.2f}")
print(f"Valor-p: {valor_p:.4f}")

# Conclusi√≥n con un nivel de significancia Œ± = 0.05
alpha = 0.05
if valor_p < alpha:
    print("Se rechaza la hip√≥tesis nula. Concluimos que el tiempo promedio de espera difiere de 8 minutos.")
else:
    print("No se puede rechazar la hip√≥tesis nula. No hay suficiente evidencia para afirmar que el tiempo promedio de espera difiere de 8 minutos.")
```

  <div style="text-align: left;"><h1>Soluci√≥n</h1></div>


<span style="color: #009929;"><strong>Inciso a)</strong></span>

El estad√≠stico de prueba \(t\) se calcula utilizando la f√≥rmula:

$$ t = \frac{(\bar{x} - \mu_0)}{(s/\sqrt{n})} $$

Donde:
- $\bar{x}$ es la media muestral,
- $\mu_0$ es la media bajo la hip√≥tesis nula,
- $s$ es la desviaci√≥n est√°ndar muestral, y
- $n$ es el tama√±o de la muestra.

En este caso:
$$ t = \frac{(17 - 18)}{(4.5/\sqrt{48})} $$

Calculamos \(t\) y obtenemos el valor del estad√≠stico de prueba.

<span style="color: #009929;"><strong>Inciso b)</strong></span>

Usaremos la tabla de distribuci√≥n t (Tabla 2 del ap√©ndice B) para encontrar los valores cr√≠ticos $t_{\text{cr√≠tico}}$ con 47 grados de libertad y $\alpha/2 = 0.025$. Luego, calcularemos el intervalo para el valor-p.

<span style="color: #009929;"><strong>Inciso c)</strong></span>

Si el valor-p es menor que $\alpha$, rechazamos la hip√≥tesis nula $(H_0)$.

<span style="color: #009929;"><strong>Inciso d)</strong></span>

Rechazamos $H_0$ si $t > t_{\text{cr√≠tico}}$ o $t < -t_{\text{cr√≠tico}}$.

<span style="color: #009929;"><strong>a)</strong></span>

$$t = \frac{(17 - 18)}{(4.5/\sqrt{48})} $$

<span style="color: #009929;"><strong>b)</strong></span>

Usando la tabla de distribuci√≥n t, encontramos $t_{\text{cr√≠tico}}$ para $\alpha/2 = 0.025$ con 47 grados de libertad.

c<span style="color: #009929;"><strong>c)</strong></span>

Compararemos el valor-p con $\alpha$ y decidiremos si rechazamos $H_0$.

<span style="color: #009929;"><strong>d)</strong></span>

Rechazamos $H_0$ si $t > t_{\text{cr√≠tico}}$ o $t < -t_{\text{cr√≠tico}}$.


<div style="color: #009929; font-weight: bold;">FIGURA 9.2 <div style="color: #5CCB5F; font-weight: bold;">Valores cr√≠ticos en la prueba de hip√≥tesis"</div></div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Generar datos de la distribuci√≥n t
x = np.linspace(-4, 4, 1000)
y = t.pdf(x, 47)

# Crear la gr√°fica
fig, ax = plt.subplots()
ax.plot(x, y, color='#009929')

# Establecer el color de fondo
ax.set_facecolor('#d4f8b7')

# Pintar el fondo externo del gr√°fico
fig.patch.set_facecolor('#D4F8B7')


# Etiquetas y texto
ax.set_xlabel('Distribuci√≥n t con 47 grados de libertad', fontsize=10)

    
# L√≠nea vertical en t cr√≠tico para alfa/2
ax.axvline(t.ppf(0.025, 47), color='#009929', linestyle='--', label=r'$t_{\alpha/2}$')

# L√≠nea vertical en -t cr√≠tico para alfa/2
ax.axvline(-t.ppf(0.025, 47), color='#009929', linestyle='--')

# Ajustes del gr√°fico
ax.legend()
ax.grid(False)
plt.show()
```

<div style="color: #009929; font-weight: bold;"><h1>Ejercicio 3</h1> </div>
Considere la prueba de hip√≥tesis siguiente.

$$H_0: \mu \geq 45 $$
$$ H_a: \mu < 45 $$

Se usa una muestra de 36. Identifique el valor-p y establezca su conclusi√≥n para cada uno de los siguientes resultados muestrales. Use $\alpha$ = 0.01 .

a) $ \bar{x} = 44$  y  $s = 5.2 $

b) $ \bar{x} = 43 $ y $ s = 4.6 $

c) $ \bar{x} = 46 $ y $ s = 5.0 $

<div style="padding: 10px; color: #009929; font-size: 1.1em;">
  <div style="text-align: left;"><h1>Solucion</h1></div>

La prueba de hip√≥tesis se realiza utilizando la f√≥rmula del valor-p para una prueba de una cola (p-value):

$$ p = P(\bar{X} < \bar{x} \,|\, H_0 \, \text{es verdadera}) $$

<span style="color: #009929;"><strong>Donde:</strong></span>

- $\bar{X}$ es la media muestral bajo la hip√≥tesis nula,
- $\bar{x}$ es la media muestral observada,
- $n$ es el tama√±o de la muestra, y
- $s$ es la desviaci√≥n est√°ndar muestral.

Si $p < \alpha$, se rechaza la hip√≥tesis nula.

<span style="color: #009929;"><strong>a)</strong></span>

 Para $\bar{x} = 44$ y $s = 5.2$:

$$ p = P(\bar{X} < 44 \,|\, \mu \geq 45) $$

<span style="color: #009929;"><strong>b)</strong></span>

 Para $\bar{x} = 43$ y $s = 4.6$:

$$ p = P(\bar{X} < 43 \,|\, \mu \geq 45) $$

<span style="color: #009929;"><strong>c)</strong></span>

 Para $\bar{x} = 46$ y $s = 5.0$:

$$ p = P(\bar{X} < 46 \,|\, \mu \geq 45) $$

<div style="color: #009929; font-weight: bold;">FIGURA 9.3 <div style="color: #5CCB5F; font-weight: bold;">Valores cr√≠ticos en la prueba de hip√≥tesis"</div></div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Par√°metros del problema
mu_0 = 45
alpha = 0.01
n = 36

# Valores cr√≠ticos para la regi√≥n de rechazo
critical_value = t.ppf(alpha, n-1)

# Datos para cada caso
sample_means = [44, 43, 46]
sample_std_devs = [5.2, 4.6, 5.0]

# Crear la gr√°fica
fig, ax = plt.subplots()

# Dibujar la distribuci√≥n t bajo H0
x = np.linspace(mu_0 - 3 * (sample_std_devs[0] / np.sqrt(n)), mu_0 + 3 * (sample_std_devs[0] / np.sqrt(n)), 1000)
ax.plot(x, t.pdf(x, n-1), color='#009929', label=r'$H_0: \mu \geq 45$')

# L√≠nea vertical en t cr√≠tico para alfa
ax.axvline(mu_0 + critical_value * (sample_std_devs[0] / np.sqrt(n)), color='#009929', linestyle='--', label=r'Regi√≥n de Rechazo')

# Pintar el √°rea de rechazo
x_fill = np.linspace(mu_0 + critical_value * (sample_std_devs[0] / np.sqrt(n)), mu_0 + 3 * (sample_std_devs[0] / np.sqrt(n)), 1000)
y_fill = t.pdf(x_fill, n-1)
ax.fill_between(x_fill, y_fill, color='#009929', alpha=0.3)

# Establecer el color de fondo
ax.set_facecolor('#d4f8b7')

# Pintar el fondo externo del gr√°fico
fig.patch.set_facecolor('#D4F8B7')

# Etiquetas y texto
ax.set_xlabel('Media Muestral ($\\bar{x}$)', fontsize=10)

# Ajustes del gr√°fico
ax.legend()
ax.grid(False)

# Mostrar la gr√°fica
plt.show()

```

## <div style="display: inline-block; background-color: #009929; padding: 8px; border-radius: 50%; color: white; font-weight: bold;">9.5</div><span style="color:#009929"> Proporcion Poblacional</span>
<div style="text-align: justify;">
En esta secci√≥n se describe c√≥mo realizar una prueba de hip√≥tesis para la proporci√≥n poblacional p si mediante p0 se denota el valor hipot√©tico para la proporci√≥n poblacional.
Las tres formas de una prueba de hip√≥tesis para la proporci√≥n poblacional son las siguientes:
</div>    
<div style="text-align: justify;">
    <div style="text-align: justify;">
        <table border="3px">
            <tr>
            <td>$H_{0}:p \geq p_0$</td>
            <td>$H_{0}:p \leq  p_0$</td>
            <td>$H_{0}:p =  p_0$</td>
        </tr>
            <tr>
            <td>$H_{a}:p < p_0$</td>
            <td>$H_{a}:p > p_0$</td>
            <td>$H_{a}:p \neq p_0$</td>
        </tr>
        </table>
    </div>
La primera forma es una prueba de cola inferior, la segunda es de cola superior y la tercera es de dos colas. 
Las pruebas de hip√≥tesis para la proporci√≥n poblacional se basan en la diferencia entre la proporci√≥n muestral p y la proporci√≥n poblacional hipot√©tica p0. Los m√©todos son semejantes a los usados para las pruebas de hip√≥tesis de la media poblacional. La diferencia radica en que para calcular el estad√≠stico de prueba se usa la proporci√≥n muestral y su error est√°ndar. Para determinar si la hip√≥tesis nula es rechazada, se utiliza el m√©todo del valor-p o el m√©todo del valor cr√≠tico. <br>         
</div>

<div style="padding: 10px; background-color: #009929; color: white; font-size: 1.1em;">
  <div style="text-align: left;"><h1>Ejemplo</h1></div>

<div style="text-align: justify;">
     Considere el caso del Coliseo de la UMSA del equipo de Voleibol. En los a√±os anteriores, 20% de los jugadores del campo eran mujeres. Un mes despu√©s se realizada la promoci√≥n, el directivo del equipo solicit√≥ un estudio estad√≠stico para determinar si la proporci√≥n de jugadoras hab√≠a aumentado. Como el objetivo es determinar si la proporci√≥n de jugadoras se increment√≥, lo apropiado es una prueba de cola superior en la que $H_{a}:p > 0,20$<br>
    Las hipotesis nula y alternativa para esta prueba son:
    <table border="3px">
            <tr>
                <td>$H_{0}:p \geq 0,20$</td>
            </tr>
            <tr>
                <td>$H_{a}:p > 0,20$</td>
            </tr>
   </table>
    Si $H_0$ se puede rechazar, los resultados de la prueba dar√°n sustento estad√≠stico a la conclusi√≥n de que la proporci√≥n de voleibolistas aument√≥ y que la promoci√≥n fue efectiva. 
    El directivo del campo especific√≥ que el nivel de significancia sera $Œ± =0,05$ para realizar esta prueba de hip√≥tesis<br>
    El paso siguiente en el procedimiento de prueba de hip√≥tesis es seleccionar una muestra y calcular el valor del estad√≠stico de prueba adecuado. 
</div>

<div style="padding: 10px; background-color: #D4F8B7; color: black; font-size: 1.1em;">
  <div style="text-align: left;">Error estandar de $\bar{p}$:</div>
  <div style="text-align: center;">\(\large  \sigma_{\bar{p}}=\sqrt{\frac{p_0(1-p_0)}{n}}\)</div>
</div>

<div style="text-align: justify;">
  <div style="text-align: left;">En el capitulo 7 se dijo que si $np \geq 5$ y $n(1-p)$ la distribucion de muestreo de $\bar {p}$ puede aproximarse mediante una distribcion normal entonces aplicamos el estadistico:</div>
  <div style="text-align: center;">\(\large z = \frac{\bar{p} - p_0}{\sigma_{\bar{p}}}\)</div>

  <div style="text-align: left;">ESTADISTICO DE PRUEBA EN LAS PRUEBAS DE HIPOTESIS PARA LA PROPORCION POBLACIONAL:</div>
  <div style="text-align: center;">\(\large z = \frac{\bar{p} - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}}\)</div>

## <div style="display: inline-block; background-color: #009929; padding: 8px; border-radius: 50%; color: white; font-weight: bold;">9.6</div><span style="color:#009929"> Prueba de hipotesis y toma de decisiones</span>

<div style="text-align: justify;">
En las secciones previas de este cap√≠tulo se estudiaron aplicaciones de pruebas de hip√≥tesis consideradas pruebas de signifi cancia. Despu√©s de formular las hip√≥tesis nula y alternativa, se selecciona una muestra y se calcula el valor de un estad√≠stico de prueba y el valor-$p$ asociado.
Se compara, entonces, el valor-p con una probabilidad controlada de cometer un error tipo I, Œ±,que se conoce como nivel de significancia para la prueba. Si el valor-$p$ ‚â§ Œ±, se concluye ‚Äúrechazar $H_0$‚Äù, y los resultados se declaran significantes; de otra manera, se concluye ‚Äúno rechazar $H_0$‚Äù. Con una prueba de significancia se controla la probabilidad de cometer un error tipo I, pero no uno tipo II. Por tanto, se recomienda la conclusi√≥n ‚Äúno rechazar $H0$‚Äù m√°s que ‚Äúaceptar $H_0$‚Äù,por que esta √∫ltima nos expone al riesgo de cometer un error tipo II de aceptar $H_0$ cuando es falsa.
Con la conclusi√≥n de ‚Äúno rechazar $H0$‚Äù la evidencia estad√≠stica se considera no concluyente y es por lo general un indicador para postergar una decisi√≥n o una acci√≥n hasta que se pueda realizar mayor investigaci√≥n y pruebas.
Pero si el prop√≥sito de una prueba de hip√≥tesis es tomar cierta decisi√≥n cuando H0 es verdadera y una decisi√≥n diferente cuando Ha es verdadera, quien debe tomarla desear√°, y en muchos casos tendr√° que actuar tanto en el caso en que la conclusi√≥n sea no rechazar $H_0$ como en el caso en que sea rechazar $H_0$. Si se da esta situaci√≥n, los expertos en estad√≠stica recomiendan controlar la probabilidad de cometer un error tipo II. Con las probabilidades controladas de cometer tanto un error tipo I como tipo II, la conclusi√≥n de la prueba de hip√≥tesis es ya sea aceptar $H_0$ o rechazar $H_0$. En el primer caso, se concluye que $H_0$ es verdadera, mientras que en el segundo, que $H_a$ es verdadera. As√≠, se puede tomar una decisi√≥n y emprender una acci√≥n apropiada cuando se lleg√≥ a una conclusi√≥n.
Una buena ilustraci√≥n de una prueba de hip√≥tesis para tomar decisiones es el muestreo de aceptaci√≥n de lotes, un tema que se discutir√° con m√°s detalle en el cap√≠tulo 20. Por ejemplo, un director de control de calidad tiene que decidir si acepta un pedido de bater√≠as de un proveedor o si lo rechaza por ser de mala calidad. Suponga que las especifi caciones de dise√±o indican que se requieren bater√≠as con una vida √∫til promedio de por lo menos 120 horas. Para evaluar si el pedido recibido satisface esta especifi caci√≥n, se selecciona una muestra de 36 bater√≠as y se prueban. Con base en esta muestra, se deber√° tomar la decisi√≥n de aceptar el pedido o devolverlo al proveedor por no tener la calidad adecuada. Sea Œº el n√∫mero medio de horas de vida √∫til que tienen las bater√≠as del env√≠o. Las hip√≥tesis nula y alternativa para la media poblacional se
presentan a continuaci√≥n.
    <center> $H_0: \mu \geq 120$ <br> 
 $H_a: \mu < 120 $</center>

Si $H_0$ es rechazada, se concluye que la hip√≥tesis alternativa es verdadera. Esta conclusi√≥n indica que lo adecuado es devolver el pedido al proveedor. Pero si $H_0$ no es rechazada, la personaque toma la decisi√≥n deber√° determinar qu√© medidas tomar. As√≠, sin haber concluido que $H_0$ es verdadera, sino s√≥lo por no haberla rechazado, dicha persona tendr√° que aceptar el env√≠o y considerarlo de la calidad adecuada.
En tales situaciones es recomendable que el procedimiento de prueba de hip√≥tesis se ampl√≠e para controlar la probabilidad de cometer un error tipo II. Como se tomar√° una decisi√≥n y alguna medida cuando H0 no sea rechazada, ser√° √∫til conocer la probabilidad de cometer un error de este tipo. En las secciones 9.7 y 9.8 se explica c√≥mo calcular la probabilidad de cometer
un error tipo II y ajustar el tama√±o de la muestra para controlar esta probabilidad.

## <div style="display: inline-block; background-color: #009929; padding: 8px; border-radius: 50%; color: white; font-weight: bold;">9.7</div><span style="color:#009929"> Calculo de la probabilidad de los errores tipo II </span>

<div style="text-align: justify;">En esta secci√≥n se describe c√≥mo calcular la probabilidad de cometer un error tipo II en una prueba de hip√≥tesis para la media poblacional. Este procedimiento se ilustra usando el ejemplo del muestreo de aceptaci√≥n de lotes descrito en la secci√≥n 9.6. Las hip√≥tesis nula y alternativa para el n√∫mero medio de horas de vida √∫til de un pedido de bater√≠as son: $(H_0: \mu \geq 120)$ y $(H_a: \mu < 120)$. Si $(H_0)$ es rechazada, la decisi√≥n ser√° devolver el producto al proveedor, debido a que la media del n√∫mero de horas de vida √∫til es menor que la especificada de 120 horas. Si $(H_0)$ no se rechaza, la decisi√≥n ser√° aceptar el pedido.
Suponga que se usa el nivel de significancia $(\alpha = 0.05)$ para realizar la prueba de hip√≥tesis.
</div>

$$ 
z = \frac{{\bar{x} - \mu_0}}{{\frac{\sigma}{\sqrt{n}}}} = \frac{{\bar{x} - 120}}{{\frac{\sigma}{\sqrt{n}}}} 
$$


<div style="text-align: justify;">Con base en el m√©todo del valor cr√≠tico y $z_{0.05} = 1.645$, la regla de rechazo en esta prueba de cola inferior es:
</div>
$$Rechazar H_0   si z \leq -1.645$$

<div style="text-align: justify;">Asuma que se seleccionar√° una muestra de 36 bater√≠as y que por pruebas anteriores se puede considerar que se conoce la desviaci√≥n est√°ndar poblacional y que su valor es $\sigma = 12$ horas. La regla de rechazo indica que $H_0$ ser√° descartada si:
</div>
$$ z = \frac{{\bar{x} - 120}}{{\frac{12}{\sqrt{36}}}} \leq -1.645 $$

<div style="text-align: justify;">Donde $\bar{x}$ es la media de la muestra, $\mu_0$ es la hip√≥tesis nula $\mu_0 = 120$, $\sigma$ es la desviaci√≥n est√°ndar poblacional $\sigma = 12$, y $n$ es el tama√±o de la muestra $n = 36$.
</div>

<div style="text-align: justify;">Al despejar $x$ de la expresi√≥n anterior, tenemos que $H_0$ ser√° rechazada si:
</div>
$$x \geq 120 + 1.645 \cdot \frac{12}{\sqrt{36}} = 116.71$$
<div style="text-align: justify;">Rechazar $H_0$ siempre que $x \geq 116.71$ significa que se tomar√° la decisi√≥n de aceptar el pedido siempre que $x \leq 116.71$.
</div>
<div style="text-align: justify;">Con esta informaci√≥n, se pueden calcular las probabilidades asociadas con cometer un error tipo II. Primero, recuerde que se comete este error cuando la verdadera media del pedido es menor de 120 horas y se decide aceptar $H_0: \mu \geq 120$. Por tanto, para calcular la probabilidad de cometerlo, se debe elegir un valor de $\mu$ menor que 120 horas.
</div>
<div style="text-align: justify;">Por ejemplo, suponga que la calidad del env√≠o es pobre si la vida promedio de las bater√≠as es $\mu = 112$ horas. Si en realidad es verdad que $\mu = 112$, ¬øcu√°l es la probabilidad de aceptar $H_0: \mu \geq 120$ y cometer as√≠ un error tipo II? Observamos que es la probabilidad de que la media muestral $x$ sea mayor de 116.71 cuando $\mu = 112$.
</div>
<div style="text-align: justify;">En la figura 9.8 se presenta la distribuci√≥n de muestreo de $x$ si la media es $\mu = 112$. El √°rea sombreada en la cola superior da la probabilidad de obtener $x \geq 116.71$. Utilizando la distribuci√≥n normal est√°ndar vemos que para $x \geq 116.71$:
</div>
$$z = \frac{x - \mu}{\frac{\sigma}{\sqrt{n}}} = \frac{116.71 - 112}{\frac{12}{\sqrt{36}}} = 2.36 $$

<div style="text-align: justify;">La tabla de probabilidad normal est√°ndar indica que para $z = 2.36$, el √°rea en la cola superior es $1.0000 - 0.9909 = 0.0091$. Entonces, $0.0091$ es la probabilidad de cometer un error tipo II cuando $\mu = 112$. Si usamos $\beta$ para denotar la probabilidad de cometer este error, tenemos que si $\mu = 112$, $\beta = 0.0091$. Podemos concluir que si la media de la poblaci√≥n es $112$ horas, la probabilidad de incurrir en un error tipo II es solo $0.0091$.
</div>

<div style="color: #009929; font-weight: bold;">FIGURA 9.8 <div style="color: #5CCB5F; font-weight: bold;">Probabilidad de un error tipo II cuando $\mu$=112</div></div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√°metros
n = 36
alpha = 0.05
critical_value = norm.ppf(alpha)

# Valor espec√≠fico de Œº
mu = 112

# Calcular la media y la desviaci√≥n est√°ndar de la distribuci√≥n de muestreo
sampling_mean = mu
sampling_std = 12 / np.sqrt(n)

# Generar datos para la distribuci√≥n de muestreo
x_values = np.linspace(mu - 3 * sampling_std, mu + 3 * sampling_std, 1000)
y_values = norm.pdf(x_values, loc=sampling_mean, scale=sampling_std)

# Calcular el valor z correspondiente a la zona de rechazo
rejection_zone = x_values[x_values >= mu + critical_value * sampling_std]

# Crear la gr√°fica
fig, ax = plt.subplots()
ax.plot(x_values, y_values, label=f'Distribuci√≥n de muestreo de $x$, Œº = {mu}', color='#009929')
ax.fill_between(rejection_zone, norm.pdf(rejection_zone, loc=sampling_mean, scale=sampling_std), alpha=0.3, label=f'Zona de rechazo, Œº = {mu}', color='#D4F8B7')

# Punto espec√≠fico en la gr√°fica
plt.scatter([116.71], [0], color='#D4F8B7', marker='o', label='Punto espec√≠fico (116.71, 0)')

# Anotaciones
plt.annotate(f'$\mu = {mu}$', xy=(mu, 0), xytext=(mu, 0.02), ha='center', arrowprops=dict(facecolor='#D4F8B7', shrink=0.05))

# Etiquetas y t√≠tulo
ax.set_title('Distribuci√≥n de muestreo de $x$ con Zona de rechazo y Punto Espec√≠fico', fontsize=14)
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('Densidad de probabilidad', fontsize=12)
ax.legend()

# Cambiar el color de fondo del gr√°fico
ax.set_facecolor('#d4f8b7')

# Cambiar el color de fondo de la figura
fig.patch.set_facecolor('#D4F8B7')

plt.show()
```

## <div style="display: inline-block; background-color: #009929; padding: 8px; border-radius: 50%; color: white; font-weight:bold;">9.8</div><span style="color:#009929;text-align: justify;">Determinaci√≥n del tama√±o de la muestra en una prueba de hip√≥tesis para la media poblacional </span>


<div style="text-align: justify;">Considera realizar una prueba de hip√≥tesis para el valor de la media poblacional. El nivel de significancia elegido por el usuario determina la probabilidad de cometer un error tipo I en esta prueba. Al controlar el tama√±o de la muestra, el usuario tambi√©n controla la probabilidad de cometer un error tipo II. A continuaci√≥n se muestra c√≥mo determinar el tama√±o de la muestra en la prueba de hip√≥tesis de cola inferior para la media poblacional:
</div>
$$H_0: \mu \leq \mu_0$$
$$H_a: \mu > \mu_0$$

<div style="text-align: justify;">En la Figura 9.8, la gr√°fica superior presenta la distribuci√≥n de muestreo de $x$ cuando $H_0$ es verdadera y $\mu > \mu_0$. En una prueba de cola inferior, el valor cr√≠tico del estad√≠stico de prueba se denota como \(c\). La l√≠nea vertical, $c$, en la gr√°fica superior de la figura, se√±ala el valor correspondiente de $x$. Observa que si $H_0$ es rechazada cuando $x \geq c$, la probabilidad de cometer un error tipo I ser√° $\alpha$. Si $z_\alpha$ representa el valor de $z$ que corresponde al √°rea $\alpha$ en la cola superior de la distribuci√≥n normal est√°ndar, la siguiente f√≥rmula se emplea para calcular $c$:
</div>
$$c = \mu_0 - z_\alpha \frac{\sigma}{\sqrt{n}}$$


<div style="text-align: justify;">Ahora, lo que buscamos es elegir un valor para \(c\), de manera que cuando $H_0$ sea rechazada y $H_a$ aceptada, la probabilidad de cometer un error tipo I sea igual a la probabilidad elegida para $\alpha$, y la probabilidad de cometer un error tipo II sea igual al valor elegido para $\beta$. Por consiguiente, con ambas ecuaciones (9.5) y (9.6) se debe obtener el mismo valor de $c$ y la ecuaci√≥n siguiente debe satisfacerse.
</div>
$$
\frac{{\mu_0 - \mu_a}}{{\frac{{z_\alpha}}{{\sqrt{n}}}}} = \frac{{z_\beta}}{{\sqrt{n}}}
$$

Para determinar el tama√±o de muestra que se necesita, primero se despeja $\sqrt{n}$ como sigue.

$$
\sqrt{n} = \frac{{(\mu_0 - \mu_a)}}{{(z_\alpha - z_\beta)}} \cdot \sigma
$$

y

$$
n = \left(\frac{{z_\alpha - z_\beta \cdot \sigma}}{{\mu_0 - \mu_a}}\right)^2
$$
<div style="text-align: justify;">Al elevar al cuadrado ambos lados de la expresi√≥n, obtenemos la f√≥rmula siguiente para el tama√±o de la muestra necesario en una prueba de hip√≥tesis de una cola para la media poblacional.
</div>
<h3 style="color:#009929"> Tama√±o de la Muestra en una Prueba de Hip√≥tesis de una Cola para la Media Poblacional </h3>


$$
n = \left(\frac{{(z_\alpha - z_\beta)^2 \cdot \sigma^2}}{{(\mu_0 - \mu_a)^2}}\right)
$$

donde:

- $z_\alpha$ = el valor de $z$ que proporciona un √°rea de $\alpha$ en la cola superior de la distribuci√≥n normal est√°ndar.
- $z_\beta$ = el valor de $z$ que proporciona un √°rea de $\beta$ en la cola superior de la distribuci√≥n normal est√°ndar.
- $\sigma$ = la desviaci√≥n est√°ndar poblacional.
- $\mu_0$ = el valor de la media poblacional en la hip√≥tesis nula.
- $\mu_a$ = el valor de la media poblacional utilizado para el error tipo II.

<div style="text-align: justify;">Nota_: Para una prueba de hip√≥tesis de dos colas, en la ecuaci√≥n (9.7) se usa $z_{\alpha/2}$ en lugar de $z_\alpha$. Aunque la l√≥gica de la ecuaci√≥n (9.7) se desarroll√≥ para la prueba de hip√≥tesis mostrada en la figura 9.10, tambi√©n es v√°lida en cualquier prueba de hip√≥tesis de una cola para la media poblacional. En una prueba de hip√≥tesis de dos colas para la media poblacional se usa $z_{\alpha/2}$ en lugar de $z_\alpha$ en la misma ecuaci√≥n.
</div>
<div style="text-align: justify;">Volviendo al ejemplo del muestreo de aceptaci√≥n de lotes presentado en las secciones 9.6 y 9.7. Las especificaciones de dise√±o para el embarque de las bater√≠as indican una vida media √∫til de por lo menos 120 horas. Los pedidos se regresan si $H_0$ es rechazada: $\mu \geq 120$. Supongamos que el gerente de control de calidad establece las siguientes declaraciones acerca de las probabilidades admisibles de cometer los errores tipo I y tipo II:
</div>
<div style="text-align: justify;"><h3 style="color:#009929"> Declaraci√≥n para el error tipo I:</h3> Si la vida media de las bater√≠as del pedido es $\mu \leq 120$, estoy dispuesto a asumir el riesgo de que la probabilidad de rechazar el embarque sea $\alpha = 0.05$.
</div>
<div style="text-align: justify;"><h3 style="color:#009929"> Declaraci√≥n para el error tipo II:</h3> Si la vida media de las bater√≠as del pedido es 5 horas por debajo de lo que indican las especificaciones (es decir, $\mu = 115$, estoy dispuesto a asumir el riesgo de que la probabilidad de aceptar el embarque sea $\beta = 0.10$.
</div>
<div style="text-align: justify;">Estas declaraciones se basan en el criterio del gerente. Otra persona podr√≠a establecer diferentes restricciones para las probabilidades. Sin embargo, las declaraciones acerca de las probabilidades admisibles de ambos errores deben establecerse antes de determinar el tama√±o de la muestra.
</div>
<div style="text-align: justify;">En el ejemplo, $\alpha = 0.05$ y $\beta = 0.10$. Mediante la distribuci√≥n de probabilidad normal est√°ndar, se tiene $z_{0.05} = 1.645$ y $z_{0.10} = 1.28$.De acuerdo con lo dicho al especificar las probabilidades para los errores, observamos que $\mu_0 = 120$ y $\mu_a = 115$. Por ultimo, supusimos que las desviaciones estandar poblacional se conocia y era $\alpha = 12$. 
</div>

$$
n = \left(\frac{{(1.645 + 1.28)^2 \cdot (12)^2}}{{(120 - 115)^2}} = 49.3\right)
$$

<div style="text-align: justify;">Como las probabilidades de los dos errores tipo I y tipo II se han controlado usando $n = 50$, queda justificado que el gerente de control de calidad utilice las declaraciones $H_0$ es aceptada o $H_0$ es rechazada en esta prueba de hip√≥tesis. Las inferencias correspondientes se hacen teniendo probabilidades admisibles de cometer un error de cualquiera de ambos tipos.
</div>

<div style="text-align: justify;">Acerca de la relaci√≥n entre $\alpha$, $\beta$ y el tama√±o $n$ de la muestra caben tres observaciones.


1. Una vez que se tienen dos de estos tres valores, el tercero puede calcularse.

2. Dado un nivel de significancia $\alpha$, aumentando el tama√±o de la muestra se reduce $\beta$.

3. Dado un tama√±o de muestra, al reducirse $\alpha$ aumenta $\beta$ y al incrementarse $\alpha$, disminuye $\beta$.
    
    
<div style="text-align: justify;">La tercera observaci√≥n debe tenerse en cuenta cuando no se controla la probabilidad de cometer un error tipo II. Dicha observaci√≥n indica que no se deben elegir niveles de significancia $\alpha$ innecesariamente peque√±os. Para un tama√±o de muestra dado, elegir un nivel de significancia peque√±o implica m√°s riesgo de cometer un error tipo II. Personas con poca experiencia piensan que al realizar una prueba de hip√≥tesis es mejor usar siempre valores peque√±os de $\alpha$, lo cual es cierto si la √∫nica preocupaci√≥n es cometer un error tipo I. Sin embargo, los valores peque√±os de $\alpha$ tienen la desventaja de incrementar la probabilidad de cometer un error tipo II.
</div>

Para resolver este ejercicio, podemos utilizar la f√≥rmula para el tama√±o de muestra en una prueba de hip√≥tesis de una cola para la media poblacional. La f√≥rmula es:

$$ n = \left(\frac{(z_\alpha + z_\beta)^2 \cdot \sigma^2}{(\mu_0 - \mu_a)^2}\right) $$

donde:

- $ z_\alpha $ es el valor cr√≠tico correspondiente al nivel de significancia $ \alpha $, y como $ \alpha = 0.05 $, entonces $ z_\alpha = -1.645 $ (ya que estamos en la cola inferior),
- $ z_\beta $ es el valor cr√≠tico correspondiente a la probabilidad de error tipo II $ \beta $, y como $\beta = 0.10 $, entonces $z_\beta = -1.28 $ (seg√∫n la informaci√≥n proporcionada),
- $ \sigma $ es la desviaci√≥n est√°ndar poblacional, que en este caso es $ \sigma = 5 $,
- $ \mu_0 $ es la media bajo la hip√≥tesis nula, que es $ \mu_0 = 10 $,
- $ \mu_a $ es la media bajo la hip√≥tesis alternativa, que es $ \mu_a = 9 $.

Sustituimos estos valores en la f√≥rmula:

$$ n = \left(\frac{(-1.645 - (-1.28))^2 \cdot 5^2}{(10 - 9)^2}\right) $$

Calculamos:

$$ n = \left(\frac{(-0.365)^2 \cdot 25}{1}\right) $$

$$ n = \left(\frac{0.133225 \cdot 25}{1}\right) $$

$$ n = \left(\frac{3.330625}{1}\right) $$

$$ n = 3.330625 $$

Redondeamos hacia arriba para obtener un n√∫mero entero, ya que el tama√±o de la muestra debe ser un n√∫mero entero positivo:

$$ n = 4 $$

Por lo tanto, se recomienda un tama√±o de muestra de 4 para reducir la probabilidad de error tipo II a $ \beta = 0.10 $ si la media poblacional verdadera es $ \mu = 9 $.


<div style="color: #009929; font-weight: bold;">FIGURA 9.8 <div style="color: #5CCB5F; font-weight: bold;">Ejercicio 1</div></div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√°metros del problema
alpha = 0.05
beta = 0.10
sigma = 5
mu_0 = 10
mu_a = 9

# Valores cr√≠ticos z_alpha y z_beta
z_alpha = norm.ppf(alpha)
z_beta = norm.ppf(beta)

# Datos para la distribuci√≥n muestral bajo H0
x_h0 = np.linspace(mu_0 - 4 * sigma, mu_0 + 4 * sigma, 1000)
y_h0 = norm.pdf(x_h0, mu_0, sigma)

# Datos para la distribuci√≥n muestral bajo Ha
x_ha = np.linspace(mu_a - 4 * sigma, mu_a + 4 * sigma, 1000)
y_ha = norm.pdf(x_ha, mu_a, sigma)

# Crear la gr√°fica
fig, ax = plt.subplots()
ax.plot(x_h0, y_h0, label=r'$H_0: \mu \geq 10$', color='green')
ax.plot(x_ha, y_ha, label=r'$H_a: \mu < 10$', color='green')

# Marcar los valores cr√≠ticos

ax.axvline(mu_a + z_beta * sigma, color='green', linestyle='--', label=r'$z_\beta$')

# Establecer el color de fondo
ax.set_facecolor('#d4f8b7')
# Pintar el fondo externo del gr√°fico
fig.patch.set_facecolor('#D4F8B7')
ax.set_xlabel(r'$\bar{x}$ (Media Muestral)', fontsize=12)

# Agregar informaci√≥n sobre los valores cr√≠ticos

ax.text(0.76, 0.15, f'$z_{{\\beta}}$ = {z_beta:.2f}', transform=ax.transAxes, fontsize=9, color='green')

# Sombrear las regiones cr√≠ticas
ax.fill_between(x_h0, y_h0, where=(x_h0 <= mu_0 + z_alpha * sigma), color='green', alpha=0.3)
ax.fill_between(x_ha, y_ha, where=(x_ha <= mu_a + z_beta * sigma), color='green', alpha=0.3)

# Leyenda y etiquetas adicionales
ax.legend()
ax.grid(False)
plt.show()
```
